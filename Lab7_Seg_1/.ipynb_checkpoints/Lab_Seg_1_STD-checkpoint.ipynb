```python
## For all requirements: The quality of the results is evaluated.


# 1- Read the image 
# [VERY IMPORTANT] and convert it to float: image = image.astype(float)

# read image golf
original_image=io.imread('./imgs/exp1/golf.jpeg')
show_images([original_image],['Original'])

# 2- Retreive the 3 channels of the image(R,G,B) (will be used in the coming steps)
red_channel=original_image[:,:,0]
green_channel=original_image[:,:,1]
blue_channel=original_image[:,:,2]

showHist(red_channel)
showHist(green_channel)
showHist(blue_channel)


show_images([red_channel,green_channel,blue_channel],['red_channel','green_channel','blue_channel'])


# 3- Visually (not by code) detect the dominating color channel (C) for the backgroung (R, G or B)
# Green

# 4- Draw the dominating channel (r, g or b) thresholded by some threshold (to make sure your choice is meaningful)
dominating_channel_thersholded=green_channel>100 #binay whrere true when >100
print(dominating_channel_thersholded)

dominating_green_channel=np.copy(green_channel)
dominating_green_channel[np.logical_not(dominating_channel_thersholded)]=0
#0 for the vlaues less than threshold
show_images([dominating_green_channel],['Domianting Channel Thersholded'])


# 5- Get an intuitive way(There is a very simple way) to make a mask that can get the pixels where the values of the dominating channels 
# have larger values than the other channels with an appropriate threshold

red_channel=red_channel.astype(float)
green_channel=green_channel.astype(float)
blue_channel=blue_channel.astype(float)

G_R=green_channel-red_channel
G_B=green_channel-blue_channel
print("green channel-red channel")
print(G_R)
print("max")
print(np.max(G_R))
print("min")
print(np.min(G_R))



print("green channel-blue channel")
print(G_B)
print("max")
print(np.max(G_B))
print("min")
print(np.min(G_B))


#Selecting from GR

GR_mask=(G_R>50)
GR_mask=np.ones(np.shape(G_B))
# print(GR_mask)

GB_mask=G_B>50

# print(GB_mask)

mask=np.logical_and(GR_mask,GB_mask)

show_images([(mask*1).astype(np.uint8)],['Mask'])

# mat[!dominating_channel_thersholded] = 0

# 6- display the binarized image




```

```python
# Get the all pixels that have RGB values close to a required color and replace their RGB values with a target color.
# path (string) :Path of img to be read
# Replace (string) :Hexcolor to be replace[Colors near to that by thershold] will also be our aiudience
# Target New color to be applied to the Replaced pixels

def ReplacePixels(original_image,replace,threshold,R_Target,G_Target,B_Target,savePath):
    # 1- Get the required RGB color (Using a color picker)
    # #485C21 (green grass)
    required_color=replace
    required_red=int((required_color)[0:2],16)
    required_green=int((required_color)[2:4],16)
    required_blue=int((required_color)[4:6],16)

#     print("Required Red",required_red)
#     print("Required Green",required_green)
#     print("Required Blue",required_blue)


    # 2- Read image
    shape=np.shape(original_image)
#     print(shape)
    # original_image=io.imread('./imgs/exp2/images.png')

#     show_images([original_image],['Original'])

    # 3- extract R, G and B channels (as float)
    red_channel=original_image[:,:,0]
    green_channel=original_image[:,:,1]
    blue_channel=original_image[:,:,2]

#     show_images([red_channel,green_channel,blue_channel],['red_channel','green_channel','blue_channel'])


    # 4- calculate differences FOR EACH CHANNEL (between the image and the required pixel value)
    #Convert to floating
    red_channel = red_channel.astype(float)
    green_channel = green_channel.astype(float)
    blue_channel = blue_channel.astype(float)

    red_channel_difference=red_channel-required_red
    green_channel_difference=green_channel-required_green
    blue_channel_difference=blue_channel-required_blue

    # 5- calculate overall distance from the given RGB color (use any appropriate distance measure, take care of the minus distances)
    #Using Ecludien distance ########
    overall_distance=np.sqrt(red_channel_difference**2+green_channel_difference**2+blue_channel_difference**2)

    # 6-  create a mask by thresholding the differences
    thresholded_img=overall_distance<threshold #Binary array with True when values rae less than the threshold
#     print(thresholded_img)
    # print(np.min(overall_distance))
    # print(np.max(overall_distance))



    # 7- In the input image, Replace the pixels of the mask with the following color
    # R = 230 , G = 90 , B=40 like orange
    #New Channels
    Result_red=np.copy(red_channel)
    Result_green=np.copy(green_channel)
    Result_blue=np.copy(blue_channel)

    Result_red[thresholded_img]=R_Target
    Result_green[thresholded_img]=G_Target
    Result_blue[thresholded_img]=B_Target

    Result=np.zeros(shape)
    Result[:,:,0]=Result_red
    Result[:,:,1]=Result_green
    Result[:,:,2]=Result_blue

    Result=Result.astype(np.uint8)
#     print(Result)
    # Result[:,:,0]=0
    # Result[:,:,1]=0
    # Result[:,:,2]=0

#     show_images([Result],['Result'])

    # 8- show the image or save it
    io.imsave(savePath,Result)
    return Result

```
